```python
import sympy as sp
import math

def taylor_cos(x, terms=6):
    result = 0
    for n in range(terms):
        term = ((-1) ** n) * (x ** (2 * n)) / math.factorial(2 * n)
        result += term
        print(f"Term {n+1}: {term}") 
    return result

x = sp.symbols('x')  
approximation = taylor_cos(x) 
```

```python
import matplotlib.pyplot as plt
import numpy as np

def taylor_cos(x, order):
    cos_approx = 0
    for n in range(order + 1):
        cos_approx += ((-1)**n * x**(2*n)) / math.factorial(2*n)
    return cos_approx

def exact_cos(x):
    return np.cos(x)

x_values = np.linspace(-2, 2, 400)

orders = [0, 2, 4, 6]

plt.figure(figsize=(10, 6))
for order in orders:
    y_values = [taylor_cos(x, order) for x in x_values]
    plt.plot(x_values, y_values, label=f"Order {order}")

plt.plot(x_values, exact_cos(x_values), label="Exact cos(x)", color='black', linestyle='--')

plt.title("Taylor Series Approximation of cos(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x_test = 1.5
exact_value = exact_cos(x_test)
errors = {}

for order in orders:
    approx_value = taylor_cos(x_test, order)
    error = abs(exact_value - approx_value)
    errors[order] = (approx_value, error)

print(f"Exact value of cos(1.5): {exact_value}")
for order in orders:
    approx_value, error = errors[order]
    print(f"Order {order}: Approximation = {approx_value}, Error = {error}")
```
