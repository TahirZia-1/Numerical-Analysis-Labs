```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def divided_diff(x, y):
    n = len(y)
    coef = np.zeros([n, n])
    coef[:,0] = y
    
    for j in range(1, n):
        for i in range(n-j):
            coef[i][j] = (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j] - x[i])
    
    return coef[0, :]
```

```python
def newton_poly(coef, x_data, x):
    n = len(x_data) - 1
    p = coef[n]
    for k in range(1, n+1):
        p = coef[n-k] + (x - x_data[n-k]) * p
    return p
```

```python
x = np.array([-5, -1, 0, 2])
y = np.array([-2, 6, 1, 3])

coef = divided_diff(x, y)

x_vals = np.linspace(-5, 2.1, 100)
y_vals = [newton_poly(coef, x, val) for val in x_vals]
```

```python
plt.plot(x_vals, y_vals, label='Polynomial Approximation')
plt.scatter(x, y, color='red', label='Data Points')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Newton\'s Divided Difference Polynomial Approximation')
plt.grid()
```
