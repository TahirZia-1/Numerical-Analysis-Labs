```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def func(x):
    return np.sin(x)

a = -np.pi
b = np.pi
k = 11


actualValue, _ = quad(func,a,b)

x_actual = np.linspace(a, b, 400)
y_actual = func(x_actual)

# For 10 intervals
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.grid()
plt.plot(x_actual, y_actual, label="cos(x)", color='black')
plt.title("Riemann Sum using 10 intervals for f(x) = cos(x)")
plt.xlabel('x')
plt.ylabel('y')

n = k - 1
delta = (b - a) / n
x = np.linspace(a, b, k)
x_mid = (x[1:] + x[:-1])/2
y_mid = func(x_mid)

riemannMid10 = delta * sum(y_mid);
errRiemannMid10 = abs(riemannMid10 - actualValue)

plt.bar(x_mid, y_mid, width=delta, align="center", alpha=0.4, color="purple", edgecolor="black")


# For 100 intervals

plt.subplot(1,2,2)
plt.grid()
plt.plot(x_actual, y_actual, label="cos(x)", color='black')
plt.title("Riemann Sum using 100 intervals for f(x) = cos(x)")
plt.xlabel('x')
plt.ylabel('y')
k= 100
n = k - 1
delta = (b - a) / n
x = np.linspace(a, b, k)
x_mid = (x[1:] + x[:-1])/2
y_mid = func(x_mid)

riemannMid100 = delta * sum(y_mid);
errRiemannMid100 = abs(riemannMid100 - actualValue)

plt.bar(x_mid, y_mid, width=delta, align="center", alpha=0.4, color="green", edgecolor="black")
plt.tight_layout()

print(f"Actual Value: {actualValue}")

print(f"Riemann Midpoint with 10 intervals: {riemannMid10}")
print(f"Error Riemann Midpoint with 10 intervals: {errRiemannMid10}")

print(f"Riemann Midpoint with 100 intervals: {riemannMid100}")
print(f"Error Riemann Midpoint with 100 intervals: {errRiemannMid100}")
```
