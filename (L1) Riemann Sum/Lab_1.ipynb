# Task 2

```python
def riemann_sum(f, a, b, n):
    dx = (b - a) / n
    x_midpoints = a + (np.arange(n) + 0.5) * dx
    f_values = f(x_midpoints)
    return np.sum(f_values) * dx

def cos_x(x):
    return np.cos(x)

a = 0
b = np.pi / 2
n = 100

result = riemann_sum(cos_x, a, b, n)
print("Riemann sum:", result)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cos_x(x):
    return np.cos(x)

x = np.linspace(-np.pi, np.pi, 400)
y = cos_x(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Graph of $cos(x)$ from -π to π')
plt.grid(True)
plt.show()
```

```python
a = -np.pi
b = np.pi 
n = 10
x_midpoints = a + (np.arange(n) + 0.5) * ((b - a) / n)
f_values = cos_x(x_midpoints)

fig, axs = plt.subplots(1)

plt.bar(x_midpoints, f_values, width=(b - a) / n, align='center', edgecolor='red', facecolor='none')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Bar Graph of Cosine Function using Riemann Sum')
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Graph of $cos(x)$ from -π to π for 10 Intervals')
plt.grid(True)


plt.show()
```
