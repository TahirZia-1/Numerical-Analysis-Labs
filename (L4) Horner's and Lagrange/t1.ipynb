```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_points = [0,2,4,6,8]
y_points = [0,5,4,2,3]
```

```python
def lagrange_basis(x, x_points, i):
    basis = 1
    for j in range(len(x_points)):
        if j != i:
            basis *= (x - x_points[j]) / (x_points[i] - x_points[j])
    return basis
```

```python
def lagrange_polynomial(x, x_points, y_points):
    polynomial = 0
    for i in range(len(y_points)):
        polynomial += y_points[i] * lagrange_basis(x, x_points, i)
    return polynomial
```

```python
x_values = np.linspace(-1, 9, 500)
y_values = [lagrange_polynomial(x, x_points, y_points) for x in x_values]
```

```python
plt.plot(x_points, y_points, 'ro', label='Given Points')
plt.plot(x_values, y_values, label='Lagrange Polynomial')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lagrange Polynomial Interpolation')
plt.legend()
plt.grid()
plt.show()
```
