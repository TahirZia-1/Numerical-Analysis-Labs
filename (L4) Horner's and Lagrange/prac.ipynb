```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lagrange(n,x,y,xp):
    yp = 0
    
    for i in range(n):
        p=1
        for j in range(n):
            if i != j:
                p *= (xp-x[j]) / (x[i]-x[j])
        yp += p * y[i]
        
    return yp

x = np.array([0, 2, 4, 6, 8])
y = np.array([0, 5, 4, 2, 3])


x_np = np.linspace(0,8,400)
y_np = np.zeros_like(x_np)

for i in range(x_np.size):
    y_np[i] += lagrange(y.size,x,y,x_np[i]) 
    
plt.plot(x_np,y_np)
plt.scatter(x,y,color='red')
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Lagrange interpolation function
def lagrange(n, x, y, xp):
    yp = 0
    for i in range(n):
        p = 1
        for j in range(n):
            if i != j:
                p *= (xp - x[j]) / (x[i] - x[j])
        yp += p * y[i]
    return yp

# Known data points
x = np.array([0, 2, 4, 6, 8])
y = np.array([0, 5, 4, 2, 3])

# Point at which we want to interpolate
xp = 1.5
yp = lagrange(len(x), x, y, xp)

# Plot the known data points and the interpolated value
plt.plot(x, y, 'o-', label='Original Data')  # Plot original data points
plt.scatter(x, y, color='red', label='Known Points')  # Known points in red

```
